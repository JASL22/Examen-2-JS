```python
coeficientes = [1, 5.5, 4.5, -9.5, -15.5, -6]
raices = np.roots(coeficientes)
raices_reales = [r.real for r in raices if r.imag == 0]
raices_reales.sort()
# Mostrar las ra√≠ces
for i in range(5):
    if i < len(raices_reales):
        print(f"x{i+1} = {raices_reales[i]}")
    else:
        print(f"x{i+1} = NaN")
```
