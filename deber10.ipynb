```python
import numpy as np
from scipy.linalg import lu, inv, solve, det
```

```python
# Literal a
A1a = np.array([[2, -3, 1], [5, 3, -1], [2, 0, 2]])
B1a = np.array([[5, 2, -3], [0, 1, -2], [1, 2, 1]])
result1a = A1a @ B1a
print("1a) Resultado:\n", result1a)

# Literal b
A1b = np.array([[2, -3, 1, 5], [3, -1, -3, 2], [0, 1, -2, 2]])
B1b = np.array([[-4, 2, 1], [0, 1, -2], [3, 0, 2], [1, 2, -1]])
result1b = A1b @ B1b
print("\n1b) Resultado:\n", result1b)

# Literal c
A1c = np.array([[4, 3, 0], [1, 0, -1], [5, 2, -4]])
B1c = np.array([[2, 3, -2], [2, -1, 3], [0, 2, -2]])
result1c = A1c @ B1c
print("\n1c) Resultado:\n", result1c)

# Literal d
A1d = np.array([[-2, 3, 0], [-4, 1, 2], [-1, 3, 0]])
B1d = np.array([[2, 0, -4], [1, 2, 0], [2, 3, -2]])
result1d = A1d @ B1d
print("\n1d) Resultado:\n", result1d)
```

```python
# Literal a
A2a = np.array([[4, 2, 6], [3, 0, 7], [2, -1, -3]])
if np.linalg.det(A2a) != 0:
    invA2a = inv(A2a)
    print("2a) Matriz no singular. Inversa:\n", invA2a)
else:
    print("2a) Matriz singular")

# Literal b
A2b = np.array([[1, 2, 0], [2, 1, -1], [3, 1, 1]])
if np.linalg.det(A2b) != 0:
    invA2b = inv(A2b)
    print("\n2b) Matriz no singular. Inversa:\n", invA2b)
else:
    print("\n2b) Matriz singular")

# Literal c
A2c = np.array([[1, 1, -1, 1], [1, 2, -4, -2], [2, 1, 1, 5], [1, 0, -2, -4]])
if np.linalg.det(A2c) != 0:
    invA2c = inv(A2c)
    print("\n2c) Matriz no singular. Inversa:\n", invA2c)
else:
    print("\n2c) Matriz singular")

# Literal d
A2d = np.array([[4, 0, 0, 0], [6, 7, 0, 0], [9, 11, 1, 0], [5, 4, 1, 1]])
if np.linalg.det(A2d) != 0:
    invA2d = inv(A2d)
    print("\n2d) Matriz no singular. Inversa:\n", invA2d)
else:
    print("\n2d) Matriz singular")
```

```python
import numpy as np
from sympy import symbols, Eq, solve, Matrix

# Definir la variable simbólica
alpha = symbols('alpha')

# Crear la matriz simbólica con sympy
A = Matrix([
    [1, -1, alpha],
    [2,  2, 1],
    [0, alpha, -1.5]
])

# Calcular el determinante
det_A = A.det()

# Resolver det(A) = 0
alpha_values = solve(Eq(det_A, 0), alpha)

print("4) Valores de α que hacen la matriz singular:", alpha_values)
```

```python
import numpy as np
from scipy.linalg import lu, solve

# Función para resolver sistemas usando LU
def solve_with_lu(A, b):
    P, L, U = lu(A)
    # Resolver Ly = Pb
    y = solve(L, P @ b)
    # Resolver Ux = y
    x = solve(U, y)
    return x

# Literal a
A7a = np.array([[2, -1, 1],
                [3, 3, 9],
                [3, 3, 5]])
b7a = np.array([-1, 0, 4])
sol7a = solve_with_lu(A7a, b7a)
print("7a) Solución:", sol7a)

# Literal b
A7b = np.array([[1.012, -2.132, 3.104],
                [-2.132, 4.096, -7.013],
                [3.104, -7.013, 0.014]])
b7b = np.array([1.984, -5.049, -3.895])
sol7b = solve_with_lu(A7b, b7b)
print("\n7b) Solución:", sol7b)

# Literal c
A7c = np.array([[2, 0, 0, 0],
                [1, 1.5, 0, 0],
                [0, -3, 0.5, 0],
                [2, -2, 1, 1]])
b7c = np.array([3, 4.5, -6.6, 0.8])
sol7c = solve_with_lu(A7c, b7c)
print("\n7c) Solución:", sol7c)

# Literal d
A7d = np.array([[2.1756, 4.0231, -2.1732, 5.1967],
                [-4.0231, 6.0000, 0, 1.1973],
                [-1.0000, -5.2107, 1.1111, 0],
                [6.0235, 7.0000, 0, -4.1561]])
b7d = np.array([17.102, -6.1593, 3.0004, 0.0000])
sol7d = solve_with_lu(A7d, b7d)
print("\n7d) Solución:", sol7d)
```
